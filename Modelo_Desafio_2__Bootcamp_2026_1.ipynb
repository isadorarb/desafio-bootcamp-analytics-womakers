Você recebeu uma lista com os títulos das palestras submetidas para um evento "Women in Data". Sua primeira tarefa é fazer uma contagem rápida de quantos títulos mencionam temas de BI.

Crie um programa que percorra a lista de títulos e conte quantos deles contêm as palavras-chave "dados", "análise" ou "inteligência" (sem diferenciar maiúsculas e minúsculas).

```python
titulos_palestras = [
    "Machine Learning para Negócios",
    "O Poder da Análise de Dados no Varejo",
    "Introdução ao SQL",
    "Business Intelligence com Power BI",
    "Carreira em Dados: Dicas para Mulheres",
    "Storytelling com Inteligência Artificial"
]
```

```python
#Execute seu código aqui

quantidade = 0
for titulo in titulos_palestras:
  if "dados" in titulo.lower() or "análise" in titulo.lower() or "inteligência" in titulo.lower():
    quantidade += 1
print("A quantidade de títulos com essas palavras é:", quantidade)
```

DESAFIO 2

Como parte de uma análise inicial, você precisa criar uma função para avaliar um conjunto de valores numéricos.

Crie uma função que receba uma lista de números e retorne:

o valor máximo

o valor mínimo

a média dos valores

```python
#Execute seu código aqui

def valores(lista):
  soma = 0
  vmin = lista[0]
  vmax = lista[0]
  for numero in lista:
    soma = soma + numero
    if numero < vmin:
      vmin = numero
    if numero > vmax:
      vmax = numero
  media = soma/len(lista)
  return media, vmax, vmin

lista = [1, 20, 35, 48, 52]
media, vmax, vmin = valores(lista)
print("Media:", media,", Valor máximo:", vmax, ", Valor mínimo:", vmin)
```

DESAFIO 4

Após uma palestra sobre "O Futuro da IA", foi feita uma enquete com a pergunta: "Qual foi o tópico mais impactante?". Os resultados foram armazenados em um dicionário onde a chave é o tópico e o valor é o número de votos.

Sua tarefa é calcular o número total de votos e exibir o resultado da enquete em porcentagens.

```python
resultado_enquete = {
    "IA Generativa": 120,
    "Ética em IA": 85,
    "IA no Mercado de Trabalho": 95,
    "Ferramentas de MLOps": 50
}
```

```python
#Execute seu código aqui
total_votos = 0
for i in resultado_enquete:
  total_votos = total_votos + resultado_enquete[i]
porcentagem_votos = []
for i in resultado_enquete:
  porcentagem_votos = (resultado_enquete[i]/total_votos)*100
  print("O tópico", i, "recebeu", porcentagem_votos, "% dos votos")
```

Antes de carregar os dados de novas inscritas em um evento para o banco de dados, você precisa limpá-los. Você recebeu uma lista de dicionários com nomes e cargos, mas os dados estão inconsistentes.

Crie um programa que percorra a lista e faça duas limpezas:

Remova espaços em branco no início e no final dos nomes.
Padronize o campo cargo para que todas as variações de "Analista de Dados" (ex: "data analyst", "Analista de dados") se tornem "Analista de Dados".

```python
inscricoes = [
    {"nome": "  Ana Clara ", "cargo": "data analyst"},
    {"nome": "Beatriz Guedes", "cargo": "Cientista de dados"},
    {"nome": "Carla Dias", "cargo": "analista de Dados"},
    {"nome": "Daniela Moura  ", "cargo": "Engenheira de dados"},
    {"nome": " Eduarda Santana  ", "cargo": "Analista de dados"}
]
```

```python
#Execute seu código aqui
for i in range (len(inscricoes)):
  inscricoes[i]["nome"] = inscricoes[i]["nome"].strip()
  if inscricoes[i]["cargo"].lower() == "data analyst" or inscricoes[i]["cargo"].lower() == "analista de dados":
    inscricoes[i]["cargo"] = "Analista de Dados"
print(inscricoes)
```
